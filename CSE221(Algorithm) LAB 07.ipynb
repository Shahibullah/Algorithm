```python
import heapq

def shortest_path():
    a, b, start, end = map(int, input().split())
    froms = list(map(int, input().split()))
    tos = list(map(int, input().split()))
    costs = list(map(int, input().split()))

    board = [[] for i in range(a + 1)]
    for i in range(b):
        board[froms[i]].append((tos[i], costs[i]))

    distance = [float('inf')] * (a + 1)
    parent = [-1] * (a + 1)
    distance[start] = 0
    bag = [(0, start)]

    while bag:
        d, point = heapq.heappop(bag)
        if d > distance[point]:
            continue
        for next_point, cost in board[point]:
            if distance[next_point] > d + cost:
                distance[next_point] = d + cost
                parent[next_point] = point
                heapq.heappush(bag, (distance[next_point], next_point))

    if distance[end] == float('inf'):
        print(-1)
        return

    print(distance[end])
    way = []
    now = end
    while now != -1:
        way.append(now)
        now = parent[now]

    print(*way[::-1])

shortest_path()
```

```python
import heapq

def solve():
    n, m, a, b = map(int, input().split())
    graph = [[] for i in range(n + 1)]
    for i in range(m):
        u, v, w = map(int, input().split())
        graph[u].append((v, w))

    def dijkstra(start):
        dist = [float('inf')] * (n + 1)
        dist[start] = 0
        heap = [(0, start)]
        while heap:
            d, u = heapq.heappop(heap)
            if d > dist[u]:
                continue
            for v, w in graph[u]:
                if dist[v] > d + w:
                    dist[v] = d + w
                    heapq.heappush(heap, (dist[v], v))
        return dist

    da = dijkstra(a)
    db = dijkstra(b)

    res = float('inf')
    meet = -1
    for i in range(1, n + 1):
        val = max(da[i], db[i])
        if val < res:
            res = val
            meet = i
        elif val == res and i < meet:
            meet = i

    if res == float('inf'):
        print(-1)
    else:
        print(res, meet)

solve()
```

```python
import heapq

def find_min_danger():
    cities, roads = map(int, input().split())
    network = [[] for i in range(cities + 1)]
    for i in range(roads):
        a, b, danger = map(int, input().split())
        network[a].append((b, danger))
        network[b].append((a, danger))

    danger_level = [float('inf')] * (cities + 1)
    danger_level[1] = 0
    queue = [(0, 1)]

    while queue:
        cur_danger, place = heapq.heappop(queue)
        if cur_danger > danger_level[place]:
            continue
        for neighbor, cost in network[place]:
            new_danger = max(cur_danger, cost)
            if new_danger < danger_level[neighbor]:
                danger_level[neighbor] = new_danger
                heapq.heappush(queue, (new_danger, neighbor))

    result = []
    for i in range(1, cities + 1):
        if danger_level[i] == float('inf'):
            result.append('-1')
        else:
            result.append(str(danger_level[i]))

    print(' '.join(result))

find_min_danger()
```

```python
import heapq

def beautiful_path():
    cities, links, start, goal = map(int, input().split())
    points = list(map(int, input().split()))
    road = [[] for i in range(cities + 1)]

    for i in range(links):
        u, v = map(int, input().split())
        road[u].append(v)

    price = [float('inf')] * (cities + 1)
    price[start] = points[start - 1]
    heap = [(price[start], start)]

    while heap:
        now_cost, now = heapq.heappop(heap)
        if now_cost > price[now]:
            continue
        for nxt in road[now]:
            if price[nxt] > now_cost + points[nxt - 1]:
                price[nxt] = now_cost + points[nxt - 1]
                heapq.heappush(heap, (price[nxt], nxt))

    if price[goal] == float('inf'):
        print(-1)
    else:
        print(price[goal])

beautiful_path()
```

```python
import heapq

def parity_path():
    n, m = map(int, input().split())
    u = list(map(int, input().split()))
    v = list(map(int, input().split()))
    w = list(map(int, input().split()))

    edges = [[] for i in range(n + 1)]
    for i in range(m):
        edges[u[i]].append((v[i], w[i]))

    dist = [[float('inf')] * 2 for i in range(n + 1)]
    dist[1][0] = dist[1][1] = 0
    heap = [(0, 1, -1)]

    while heap:
        d, now, parity = heapq.heappop(heap)
        for to, cost in edges[now]:
            p = cost % 2
            if p != parity:
                if d + cost < dist[to][p]:
                    dist[to][p] = d + cost
                    heapq.heappush(heap, (dist[to][p], to, p))

    ans = min(dist[n])
    print(ans if ans != float('inf') else -1)

parity_path()
```

```python
import heapq

def second_shortest_path():
    n, m, s, d = map(int, input().split())
    g = [[] for i in range(n + 1)]
    for i in range(m):
        a, b, c = map(int, input().split())
        g[a].append((b, c))
        g[b].append((a, c))

    dist = [[float('inf')] * 2 for i in range(n + 1)]
    dist[s][0] = 0
    heap = [(0, s, 0)]

    while heap:
        now_cost, now_node, used = heapq.heappop(heap)
        for to, cost in g[now_node]:
            if now_cost + cost < dist[to][0]:
                dist[to][1] = dist[to][0]
                dist[to][0] = now_cost + cost
                heapq.heappush(heap, (dist[to][0], to, 0))
                heapq.heappush(heap, (dist[to][1], to, 1))
            elif dist[to][0] < now_cost + cost < dist[to][1]:
                dist[to][1] = now_cost + cost
                heapq.heappush(heap, (dist[to][1], to, 1))

    ans = dist[d][1]
    print(ans if ans != float('inf') else -1)

second_shortest_path()
```
